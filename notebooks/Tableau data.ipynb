# Reading the DataSet

```python
df = pd.read_csv('accepted_2007_to_2018Q4.csv',low_memory=False) #entire file is read into the memeory at once
```

```python
#specified columns
filtered_df = df[['addr_state', 'grade', 'issue_d', 'loan_status', 'purpose', 'zip_code', 'annual_inc', 'int_rate', 'loan_amnt']]
```
